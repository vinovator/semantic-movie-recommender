## Clearn and Prepare data

```python
# --- Clean title.basics ---

# Replace "\N" with NaN (Not a Number) for easier handling.
basics_df.replace({"\\N": np.nan}, inplace=True)

# Filter for "movie" type only
movies_df = basics_df[basics_df["titleType"] == "movie"].copy()

# startYear and runtimeMinutes are currently "objects". convert to numeric types, errors="coerce", turns failure to NaNs
movies_df["startYear"] = pd.to_numeric(movies_df["startYear"], errors="coerce")
movies_df["runtimeMinutes"] = pd.to_numeric(movies_df["runtimeMinutes"], errors="coerce")
```

```python
# --- Clean title.principals ---

# We only care about actor, actress
principals_df = principals_df[principals_df["category"].isin(["actor", "actress"])]
principals_df.head()
```

```python
# --- Clean Kaggle plots data ---

plots_df.rename(columns = {"Title" : "primaryTitle", "Release Year" : "startYear", "Plot": "plot"}, inplace=True)
plots_df.head()
```

## Create Director and Actor Lookups: We need to map movie IDs to list of names

```python
# 1. Create a clean, fast dictionary for ID-to-Name mapping
names_df.dropna(subset=['nconst'], inplace=True)
names_df['nconst'] = names_df['nconst'].astype(str).str.strip()
name_map = names_df.set_index('nconst')['primaryName'].to_dict()
```

```python
names_df.head()
```

```python
# --- FIX: Ensure tconst is a column before setting it as the index ---
# This makes the code robust, handling cases where tconst might already be the index.
if 'tconst' not in crew_df.columns:
    crew_df.reset_index(inplace=True)
if 'tconst' not in principals_df.columns:
    principals_df.reset_index(inplace=True)
```

```python
# Clean the source DataFrames BEFORE any processing
crew_df['directors'] = crew_df['directors'].fillna('').astype(str)
principals_df['nconst'] = principals_df['nconst'].fillna('').astype(str)
```

```python
# 2. Define a reusable function to process the crew/cast
def get_names_from_ids(series, name_mapping):
    """
    Takes a Series of comma-separated ID strings, maps them to names,
    and returns a Series of comma-separated name strings.
    """
    names = (
        series
        .str.split(',')
        .explode()
        .str.strip()
        .map(name_mapping)
    )
    return names.groupby(level=0).apply(lambda x: ', '.join(x.dropna().unique()))
```

```python
# 3. Process Directors and Actors
# Set tconst as the index ONCE for both operations
crew_df.set_index('tconst', inplace=True)
principals_df.set_index('tconst', inplace=True)

# Apply the function to the now-clean directors column
directors_agg = get_names_from_ids(crew_df['directors'], name_map).to_frame(name='directors')
```

```python
directors_agg.head(20)
```

```python
# Apply the function to the top 10 actors per movie
top_actors_ids = principals_df.sort_values('ordering').groupby(level=0).head(10)['nconst']
actors_agg = get_names_from_ids(top_actors_ids, name_map).to_frame(name='actors')
```

```python
actors_agg.head(20)
```

```python
# --- UPDATED SCRIPT ---

# Create director mapping
# --- FIX: Fill missing values and ensure the column is a string type ---
crew_df['directors'] = crew_df['directors'].fillna('').astype(str)
crew_df['directors'] = crew_df['directors'].str.split(',')
directors_map = crew_df.explode('directors').rename(columns={'directors': 'nconst'})

# Clean up and align data types before merging
directors_map.dropna(subset=['nconst'], inplace=True)
names_df.dropna(subset=['nconst'], inplace=True)
directors_map['nconst'] = directors_map['nconst'].astype(str)
names_df['nconst'] = names_df['nconst'].astype(str)

directors_map = directors_map.merge(names_df[['nconst', 'primaryName']], on='nconst', how='left')
directors_agg = directors_map.groupby('tconst')['primaryName'].apply(lambda x: ', '.join(x.dropna())).reset_index(name='directors')
```

```python
directors_agg.head(25)
```

```python
# Create actor mapping (limit to top 10 for simplicity)
principals_df['ordering'] = pd.to_numeric(principals_df['ordering'])
actors_map = principals_df.sort_values('ordering').groupby('tconst').head(10)

# Apply the same data type fix here for robustness
actors_map.dropna(subset=['nconst'], inplace=True)
actors_map['nconst'] = actors_map['nconst'].astype(str)

actors_map = actors_map.merge(names_df[['nconst', 'primaryName']], on='nconst', how='left')
actors_agg = actors_map.groupby('tconst')['primaryName'].apply(lambda x: ', '.join(x.dropna())).reset_index(name='actors')
```

```python
actors_agg.head(25)
```

## Merge into a Master DataFrame

```python
# Merge IMDb data
master_df = movies_df.merge(ratings_df, on='tconst', how='left')
master_df = master_df.merge(directors_agg, on='tconst', how='left')
master_df = master_df.merge(actors_agg, on='tconst', how='left')

# Clean up columns before final merge
master_df.dropna(subset=['startYear'], inplace=True)
master_df['startYear'] = master_df['startYear'].astype(int)
```

```python
# Filter master_df to only include movies in our English ID set
master_df = master_df[master_df['tconst'].isin(english_movie_ids)]
print(f"Filtered to {len(master_df)} English movies in master IMDb list.")
```

```python
# Merge with Kaggle plot data on title and year
# This is a key step where data loss can occur if titles/years don't match perfectly
final_df = pd.merge(master_df, plots_df[['primaryTitle', 'startYear', 'plot']], on=['primaryTitle', 'startYear'], how='inner')

# Final cleanup
final_df.drop_duplicates(subset=['tconst'], inplace=True)
final_df.dropna(subset=['plot'], inplace=True)
```

## Save the processed Data

```python
# Select and reorder final columns for clarity
final_cols = [
    'tconst', 'primaryTitle', 'startYear', 'runtimeMinutes', 'genres',
    'averageRating', 'numVotes', 'directors', 'actors', 'plot'
]
final_df = final_df[final_cols]

# Save to a parquet file
# Parquet Format: A highly efficient, column-oriented data format. 
# Unlike CSVs (row-oriented), Parquet is often faster to read and takes up less space because it can apply better compression, 
# especially when columns have similar data. It also preserves data types, so you don't have to re-convert numbers 
# every time you load the file.

output_path = '../data/processed/movies_master_with_plots.parquet'
final_df.to_parquet(output_path, index=False)

print(f"Master file saved to {output_path}. Shape: {final_df.shape}")
```
